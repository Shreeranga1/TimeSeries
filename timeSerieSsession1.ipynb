```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Shreeranga1/TimeSeries/main/Petrol.csv')
```

```python
df.head()
```

```python
date1=pd.date_range(start='1/1/2001',end='30/9/2013',freq='Q')
```

```python
date1
```

```python
df['TimeStamp']=date1
```

```python
df.drop(['Year','Quarter'],inplace=True,axis=1)
```

```python
df.head()
```

```python
df.index=df.TimeStamp
```

```python
df.head()
```

```python
df.drop(columns=['TimeStamp'],inplace=True)
```

```python
df.plot()
```

Decompose and check 

```python
# handle the missing data 
```

```python
daily=df.resample('Q').mean()
```

```python
df=daily.interpolate(method='linear')
```

```python
df.plot()
```
